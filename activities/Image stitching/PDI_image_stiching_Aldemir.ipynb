#imports

```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import cv2
```

#Declarando imagems

```python
ic1 = cv2.imread('/content/drive/MyDrive/Imagens - PDI/image stitching/ic1.jpg')
ic2 = cv2.imread('/content/drive/MyDrive/Imagens - PDI/image stitching/ic2.jpg')
ic3 = cv2.imread('/content/drive/MyDrive/Imagens - PDI/image stitching/ic3.jpg')
ic4 = cv2.imread('/content/drive/MyDrive/Imagens - PDI/image stitching/ic4.jpg')
```

```python
ic1 = cv2.cvtColor(ic1, cv2.COLOR_BGR2RGB)

ic2 = cv2.cvtColor(ic2, cv2.COLOR_BGR2RGB)

ic3 = cv2.cvtColor(ic3, cv2.COLOR_BGR2RGB)

ic4 = cv2.cvtColor(ic4, cv2.COLOR_BGR2RGB)

images = np.array([ic1,ic2,ic3,ic4])
```

#Imagens

```python
#unstitched images
fig, ax = plt.subplots(1, len(images), figsize=(5 * len(images), 5))

for i in range(len(images)):
    ax[i].imshow(images[i])

plt.show()
```

#Stitch

```python
stitcher = cv2.Stitcher_create()
_bool_, stitchedImage = stitcher.stitch(images)

#cv2.imwrite('stiched-IC.png',stitchedImage)
```

#Resultado

```python
figure(figsize=(3 * len(images), 3))
plt.imshow(stitchedImage)
plt.show()
```

#Removendo bordas

```python
uncroppedIC = np.array(stitchedImage)
croppedIC = uncroppedIC[300:2650,100:7600]

#cv2.imwrite('stiched-IC-cropped.png',croppedIC)
```

#Resultado após remoção de bordas

```python
#result
figure(figsize=(3 * len(images), 3))
plt.imshow(croppedIC)
plt.show()
```
